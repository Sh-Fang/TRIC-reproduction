```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python

with open("./PerValidUpdateTimeConsume.txt",'r') as f:
    data = f.read()
    data = data.split()

data = np.array(data)
data = data.astype(np.float64) 
print(data.min())
print(data.max())
```

```python
fig, ax = plt.subplots()
plt.figure(1)
plt.plot(data)
plt.show()
fig.savefig('./PerValidUpdateTimeConsume.png',dpi=400)
```
